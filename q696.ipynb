```python
from typing import *
```

```python
class Solution:
    def countBinarySubstrings(self, s: str) -> int:
        n = len(s)
        count = 0
        for idx in range(n-1):
            if s[idx] == s[idx+1]:
                continue

            for ct in range(idx+1):
                if idx-ct < 0 or s[idx-ct] != s[idx]:
                    break

                if idx+1+ct >= n or s[idx+1+ct] != s[idx+1]:
                    break

                count += 1

        return count
```

```python
print(Solution().countBinarySubstring("00110011"))
```
