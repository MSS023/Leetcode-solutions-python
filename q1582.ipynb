```python
from typing import *
```

```python
class Solution:
    def numSpecial(self, mat: List[List[int]]) -> int:
        m, n = len(mat), len(mat[0])

        result = 0
        for r in range(m):
            count1 = mat[r].count(1)
            if count1 != 1:
                continue
            idx = mat[r].index(1)

            col = [arr[idx] for arr in mat]

            if col.count(1) != 1:
                continue

            result += 1
        
        return result
```
