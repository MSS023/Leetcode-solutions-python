```python
from typing import *
```

```python
class Solution:
    def shortestPath(self, grid: List[List[int]], k: int) -> int:
        m = len(grid)
        n = len(grid[0])

        def recursive(x, y, k, recStack):
            if (x, y) in recStack:
                return -1
            if (x, y) == (m-1, n-1):
                return 0
            minm = 1601
            if x > 0:
                if grid[x-1][y] == 0 or k > 0:
                    temp = recursive(x-1, y, k - grid[x-1][y], recStack.union(set([(x,y)])))
                    if temp > -1:
                        minm = min(minm, temp+1)
            if x < m-1:
                if grid[x+1][y] == 0 or k > 0:
                    temp = recursive(x+1, y, k - grid[x+1][y], recStack.union(set([(x,y)])))
                    if temp > -1:
                        minm = min(minm, temp+1)
            if y > 0:
                if grid[x][y-1] == 0 or k > 0:
                    temp = recursive(x, y-1, k - grid[x][y-1], recStack.union(set([(x,y)])))
                    if temp > -1:
                        minm = min(minm, temp+1)
            if y < n-1:
                if grid[x][y+1] == 0 or k > 0:
                    temp = recursive(x, y+1, k - grid[x][y+1], recStack.union(set([(x,y)])))
                    if temp > -1:
                        minm = min(minm, temp+1)
            return minm if minm != 1601 else -1
        return recursive(0, 0, k, set())
```
