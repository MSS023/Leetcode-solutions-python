```python
from typing import *
import bisect
```

```python
class Solution:
    def nextGreatestLetter(self, letters: List[str], target: str) -> str:
        idx = bisect.bisect_right(letters, target)
        if idx == len(letters):
            return letters[0]
        if letters[idx] > target:
            return letters[idx]
        return letters[idx+1]
```

```python
print(Solution().nextGreatestLetter(letters = ["c","f","j"], target = "a"))
```
