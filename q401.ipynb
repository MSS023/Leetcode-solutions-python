```python
from typing import *
```

```python
class Solution:
    def getTimeStr(self, hr: int, min: int) -> str:
        return f"{hr}:"+str(min).rjust(2,"0")

    def readBinaryWatch(self, turnedOn: int) -> List[str]:
        res = []
        for hr in range(12):
            bStr = f"{hr:b}"
            count1 = bStr.count("1")
            if count1 > turnedOn:
                continue
            minCount = turnedOn-count1

            for min in range(60):
                bStr = f"{min:b}"
                count1 = bStr.count("1")
                if count1 != minCount:
                    continue

                res.append(self.getTimeStr(hr,min))

        return res
```

```python
print(Solution().readBinaryWatch(8))
```
