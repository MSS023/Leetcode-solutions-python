```python
from typing import *
```

```python
class Solution:
    def incrementVal(m: Dict[int, int], k: int):
        if k not in m:
            m[k] = 0
        m[k] += 1

    def longestBalanced(self, nums: List[int]) -> int:
        maxm = 0
        n = len(nums)
        for idx1 in range(n-1):
            evenSet = {}
            oddSet = {}
            if nums[idx1] & 1 == 0:
                Solution.incrementVal(evenSet, nums[idx1])
            else:
                Solution.incrementVal(oddSet, nums[idx1])
            for idx2 in range(idx1+1, n):
                num = nums[idx2]

                if num & 1 == 0:
                    Solution.incrementVal(evenSet, num)
                else:
                    Solution.incrementVal(oddSet, num)

                if len(evenSet) != len(oddSet):
                    continue

                maxm = max(maxm, idx2-idx1+1)
        return maxm
```
