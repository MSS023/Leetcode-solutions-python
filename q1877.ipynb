```python
from typing import *
```

```python
class Solution:
    def minPairSum(self, nums: List[int]) -> int:
        nums.sort()
        maxm = 0
        for i in range(len(nums)//2):
            maxm = max(maxm, nums[i]+nums[-i-1])
        return maxm
```

```python
print(Solution().minPairSum([3,5,2,3]))
```
