```python
from typing import *
```

```python
class Solution:
    def numSteps(self, s: str) -> int:
        count = 0
        while len(s) > 0:
            if s == "1":
                break
            if s[-1] == "0":
                count += 1
                s = s[:len(s)-1]
                continue
            idx = len(s)-2
            while idx >= 0:
                if s[idx] == "1":
                    s = s[:idx] + "0" + s[idx+1:]
                    idx -= 1
                    continue
                s = s[:idx] + "1" + s[idx+1:]
                break
            s = s[:len(s)-1] + "0"
            count += 1
        return count
```

```python
print(Solution().numSteps("1101"))
```
