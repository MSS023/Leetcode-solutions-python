```python
from typing import *
```

```python
class Solution:
    def minSwaps(self, grid: List[List[int]]) -> int:
        m,n = len(grid), len(grid[0])

        max1s = [0]*m
        
        for r in range(m):
            for c in range(n-1, -1, -1):
                if grid[r][c] == 1:
                    max1s[r] = c
                    break

        sortedMax1s = sorted(max1s)

        for idx in range(m):
            if sortedMax1s[idx] > idx:
                return -1
            
        shouldReturn = True
        for idx in range(m):
            if max1s[idx] > idx:
                shouldReturn = False
                break
        
        if shouldReturn:
            return 0
        
        result = 0
        for idx in range(m):
            if max1s[idx] <= idx:
                continue
            for idx2 in range(idx, m):
                if max1s[idx2] <= idx:
                    for idx3 in range(idx2, idx, -1):
                        max1s[idx3], max1s[idx3-1] = max1s[idx3-1], max1s[idx3]
                        result += 1
                    break
        return result
```
