```python
from typing import *
```

```python
class Solution:
    def isTrionic(self, nums: List[int]) -> bool:
        n = len(nums)
        p = 0
        for idx in range(1,n):
            if nums[idx] <= nums[idx-1]:
                p = idx - 1
                break
        if p == 0:
            return False
        q = p
        for idx in range(p+1,n):
            if nums[idx] >= nums[idx-1]:
                q = idx - 1
                break
        if p == q or q == n-1:
            return False
        for idx in range(q+1,n):
            if nums[idx] <= nums[idx-1]:
                return False
        return True
```

```python
print(Solution().isTrionic([1,3,5,4,2,6]))
# print(Solution().isTrionic([2,4,3,3]))
```
