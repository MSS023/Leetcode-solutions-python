```python
from typing import *
import math
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

    def __str__(self):
        return f"left: {self.left}, {self.val}, right: {self.right}"
```

```python
# Definition for a binary tree node.
class Solution:
    def traverse(self, root: Optional[TreeNode]) -> List[Optional[TreeNode]]:
        arr = []
        if root is None:
            return arr

        arr = self.traverse(root.left) + arr
        arr.append(root)
        arr = arr + self.traverse(root.right)
        root.left = None
        root.right = None
        return arr

    def makeTree(self, li: List[Optional[TreeNode]]) -> Optional[TreeNode]:
        n = len(li)
        if n == 0:
            return None
        if n == 1:
            return li[0]

        mid = n//2

        root = li[mid]
        root.left = self.makeTree(li[:mid])
        root.right = self.makeTree(li[mid+1:])
        return root

    def balanceBST(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        li = self.traverse(root)
        return self.makeTree(li)
        
```

```python
obj = TreeNode(1, None, TreeNode(2,None, TreeNode(3, None, TreeNode(4,None,None))))
print(Solution().balanceBST(obj))
```
