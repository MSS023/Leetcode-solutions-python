```python
from typing import *
```

```python
class Solution:
    def findDiff(self, s1: str, s2: str) -> int:
        count = 0
        for ch1,ch2 in zip(s1,s2):
            count += 1 if ch1 != ch2 else 0
        return count

    def minOperations(self, s: str) -> int:
        final1 = "".join(["0" if idx & 1 == 0 else "1" for idx in range(len(s))])
        final2 = "".join(["1" if idx & 1 == 0 else "0" for idx in range(len(s))])
        return min(self.findDiff(final1, s), self.findDiff(final2, s))
```

```python
print(Solution().minOperations("0100"))
```
