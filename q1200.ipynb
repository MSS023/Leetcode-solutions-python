```python
from typing import *
```

```python
class Solution:
    def minimumAbsDifference(self, arr: List[int]) -> List[List[int]]:
        arr.sort()
        minm = float('inf')
        res = []
        for i in range(len(arr)-1):
            if arr[i+1]-arr[i] > minm:
                continue

            if arr[i+1]-arr[i] == minm:
                res.append([arr[i],arr[i+1]])
                continue
            
            res = [[arr[i],arr[i+1]]]
            minm = arr[i+1]-arr[i]
        
        return res
```

```python
print(Solution().minimumAbsDifference([4,2,1,3]))
```
