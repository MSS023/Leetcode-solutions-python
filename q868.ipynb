```python
from typing import *
```

```python
class Solution:
    def binaryGap(self, n: int) -> int:
        binary = f"{n:b}"
        maxm = 0

        for idx1 in range(len(binary)):
            if binary[idx1] != "1":
                continue
            
            for idx2 in range(idx1+1,len(binary)):
                if binary[idx2] != "1":
                    continue
                else:
                    maxm = max(idx2-idx1, maxm)
                    break
        return maxm
```

```python
print(Solution().binaryGap(22))
```
