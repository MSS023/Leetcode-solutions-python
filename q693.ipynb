```python
from typing import *
```

```python
class Solution:
    def hasAlternatingBits(self, n: int) -> bool:
        bits = f"{n:b}"

        curr = "1" if bits[0] == "0" else "0"
        for bit in bits:
            if bit == curr:
                return False
            curr = bit
        return True
```

```python
print(Solution().hasAlternatingBits(5))
```
