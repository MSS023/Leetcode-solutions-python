```python
from typing import *
```

```python
class Solution:
    def generateGraph(self, n: int, edges: List[List[int]]) -> List[List[int]]:
        graph: List[List[int]] = [[-1]*n for _ in range(n)]
        for u,v,w in edges:
            graph[u][v] = w
        return graph
    
    def traverse(self, node: int) -> int:
        if node == len(self.graph)-1:
            return 0
        minm = float('inf')
        for v,w in enumerate(self.graph[node]):
            if w < 0:
                continue
            minm = min(minm,self.traverse(v) + w)
        return minm

    def minCost(self, n: int, edges: List[List[int]]) -> int:
        graph = self.generateGraph(n, edges)
        self.graph = graph
        minm = float('inf')
        for u,v,_ in edges:
            graph[u][v],graph[v][u] = 2*graph[v][u], 2*graph[u][v]
            minm = min(minm, self.traverse(0))
            graph[u][v],graph[v][u] = graph[v][u]//2,graph[u][v]//2
        return minm
```

```python
print(Solution().minCost(n = 4, edges = [[0,1,3],[3,1,1],[2,3,4],[0,2,2]]))
```
