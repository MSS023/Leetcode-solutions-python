```python
from typing import *
import heapq
```

```python
class Solution:

    def __init__(self):
        self.memo = {}

    def createAdjDict(self, original: List[str], changed: List[str], cost: List[int]) -> Dict[str,Dict[str,int]]:
        adj = {}
        for o,ch,c in zip(original, changed, cost):
            if o not in adj:
                adj[o] = {}
                adj[o][o] = 0
            if ch not in adj[o]:
                adj[o][ch] = c
            adj[o][ch] = min(adj[o][ch],c)
        return adj

    def dijkstras(self, src: str, adj: Dict[str,Dict[str, int]]) -> Dict[str, int]:
        dist = {}

        dist[src] = 0
        queue = []
        heapq.heappush(queue, (0,src))
        while queue:
            w, u = heapq.heappop(queue)

            if w > dist[u]:
                continue

            if u not in adj:
                continue

            for v, c in adj[u].items():
                if v not in dist or dist[u] + c < dist[v]:
                    dist[v] = dist[u] + c
                    heapq.heappush(queue, (dist[v], v))

        return dist

    def minimumCost(self, source: str, target: str, original: List[str], changed: List[str], cost: List[int]) -> int:
        adj = self.createAdjDict(original, changed, cost)

        result = 0
        for s,t in zip(source, target):
            if s in self.memo:
                dist = self.memo[s]
            else:
                dist = self.dijkstras(s, adj)
                self.memo[s] = dist
            if t not in dist or dist[t] == float('inf'):
                return -1
            
            result += dist[t]
        return result
```
