```python
from typing import *
```

```python
class Solution:
    def prefixesDivBy5(self, nums: List[int]) -> List[bool]:
        rep = 0
        res = []
        for num in nums:
            rep = rep << 1
            rep += num
            if rep % 5 == 0:
                res.append(True)
            else:
                res.append(False)
        return res
```

```python
print(Solution().prefixesDivBy5([0,1,1]))
```
